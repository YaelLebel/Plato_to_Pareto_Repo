```python
import numpy as np
import pandas as pd
```

```python
df_maxs = pd.read_csv('../df_maxs.csv',index_col='Unnamed: 0')
df_parent_infection = pd.read_csv('parent.infections.summary.csv').dropna().reset_index(drop=True)

df_maxs_parents = df_maxs[df_maxs['Strain']!='DO']
for col in df_parent_infection.columns[1:]:
    new_col = []
    for i in range(len(df_maxs_parents)):
        this_strain = df_maxs_parents.iloc[i]['Strain']
        if this_strain=='C57BL/6J':
            this_strain = 'C57BL/2'
        elif this_strain=='NOD/ShiLtJ':
            this_strain= 'NOD/SjiLtJ'
        value = df_parent_infection[df_parent_infection['Strain']==this_strain][col].values
        if len(value)>0:
            new_col.append(value[0])
        else:
            new_col.append(np.nan)
    df_maxs_parents[col] = new_col
    
df_maxs_parents.to_csv('../df_max_parents_other_infections.csv')    
```
